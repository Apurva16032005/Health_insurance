```python
# %%
# 1. IMPORT LIBRARIES
import os
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

# Try importing fpdf, handle if missing
try:
    from fpdf import FPDF
except ImportError:
    print("âš ï¸ FPDF library not found. Please run: pip install fpdf2")
    raise

# Ensure reports directory exists
if not os.path.exists("reports"):
    os.makedirs("reports")

# %%
# 2. DEFINE REPORT CLASS
class FraudReport(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 15)
        self.cell(0, 10, 'Insurance Fraud Detection Report', 0, 1, 'C')
        self.ln(10)

    def footer(self):
        self.set_y(-15)
        self.set_font('Arial', 'I', 8)
        self.cell(0, 10, f'Page {self.page_no()}/{{nb}} - AI Sentinel System', 0, 0, 'C')

    def chapter_title(self, label):
        self.set_font('Arial', 'B', 12)
        self.set_fill_color(230, 230, 230)  # Light Gray
        self.cell(0, 6, f'  {label}', 0, 1, 'L', 1)
        self.ln(4)

    def chapter_body(self, body, color=(0,0,0)):
        self.set_font('Arial', '', 11)
        self.set_text_color(*color)
        self.multi_cell(0, 5, body)
        self.set_text_color(0, 0, 0) # Reset
        self.ln()

# %%
# 3. MOCK DATA & VISUALS
current_time = datetime.now().strftime("%Y-%m-%d %H:%M")
claim_data = {
    "claim_id": "CLM-2024-001",
    "date": current_time,
    "policy_holder": "Amit Sharma",
    "amount": "INR 1,50,000",
    "fraud_score": 0.92,
    "anomalies": [
        "Metadata traces of 'Adobe Photoshop' found.",
        "CNN detected 98% probability of pixel tampering.",
        "Claim amount exceeds hospital average for 'Viral Fever'."
    ]
}

def create_charts():
    # Heatmap
    plt.figure(figsize=(4, 3))
    plt.imshow(np.random.rand(10, 10), cmap='hot', interpolation='nearest')
    plt.title("Digital Tampering Heatmap")
    plt.axis('off')
    plt.tight_layout()
    plt.savefig("reports/temp_heatmap.png")
    plt.close()

    # Bar Chart
    plt.figure(figsize=(5, 3))
    factors = ['Metadata', 'Visuals', 'NLP', 'Amount']
    scores = [0.9, 0.85, 0.2, 0.6]
    plt.barh(factors, scores, color=['red', 'red', 'green', 'orange'])
    plt.title("Risk Factor Breakdown")
    plt.xlim(0, 1)
    plt.tight_layout()
    plt.savefig("reports/temp_chart.png")
    plt.close()

create_charts()
print("âœ… Visuals created.")

# %%
# 4. GENERATE PDF
def generate_pdf(data):
    pdf = FraudReport()
    pdf.alias_nb_pages()
    pdf.add_page()
    
    # Summary
    pdf.chapter_title("1. Claim Summary")
    pdf.set_font('Arial', '', 11)
    pdf.cell(0, 8, f"Claim ID: {data['claim_id']}", ln=1)
    pdf.cell(0, 8, f"Date: {data['date']}", ln=1)
    pdf.cell(0, 8, f"Policy Holder: {data['policy_holder']}", ln=1)
    pdf.cell(0, 8, f"Amount: {data['amount']}", ln=1)
    pdf.ln(5)
    
    # Verdict
    pdf.chapter_title("2. AI Analysis")
    if data['fraud_score'] > 0.7:
        pdf.chapter_body(f"CRITICAL RISK ({data['fraud_score']*100:.1f}%)", (200, 0, 0))
    else:
        pdf.chapter_body(f"LOW RISK ({data['fraud_score']*100:.1f}%)", (0, 150, 0))
        
    pdf.set_font('Arial', 'B', 11)
    pdf.cell(0, 8, "Anomalies:", ln=1)
    pdf.set_font('Arial', '', 10)
    for anomaly in data['anomalies']:
        pdf.cell(0, 6, f" - {anomaly}", ln=1)
    pdf.ln(10)
    
    # Images
    pdf.chapter_title("3. Forensic Evidence")
    if os.path.exists("reports/temp_heatmap.png"):
        pdf.image("reports/temp_heatmap.png", w=80)
    if os.path.exists("reports/temp_chart.png"):
        pdf.image("reports/temp_chart.png", w=100)
        
    filename = f"reports/Report_{data['claim_id']}.pdf"
    pdf.output(filename)
    return filename

# Run
try:
    report = generate_pdf(claim_data)
    print(f"ðŸŽ‰ Report generated: {report}")
except Exception as e:
    print(f"âŒ Error: {e}")
```
